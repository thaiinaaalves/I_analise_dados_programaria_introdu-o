```python
#consultando as variáveis existentes na coluna de interesse
df['GENERO']
```

```python
#filtrando pela variável desejada
df[df['GENERO']=='Feminino']
```

```python
#outra forma de filtrar
df[df['GENERO']!='Masculino']
```

```python
#filtrando por palavras, nesse caso pelas respostas que contenham não
df[df['GENERO'].str.contains('não', na=False)]
```

```python
#filtrando por idades igual ou maiores que 30 anos
df[df['IDADE']>=30]
```

```python
#filtrando por mais de uma opção.
#OBS.Podemos usar '&' para estabelecer mais uma condição a nossa consulta, ou '|' no sentido de ou
df[(df['IDADE']>=30) & (df['GENERO']=='Feminino')]
```

```python
#consultando as variáveis existentes na coluna de interesse
#e usando a função value_counts(), para fazer p calculo da quantidade de ocorrencias da variável
df['COR/RACA/ETNIA'].value_counts()
```

```python
#agora o filto deverá demonstrar a quantidade de pessoas que se declaram amarelas e com idade menor ou igual a 40 anos.
df[(df['COR/RACA/ETNIA']=='Amarela') & (df['IDADE']<=40)]
```

```python
#a função groupby agrupa dados por uma coluna específica. Ex:coluna'GENERO', e como parametro para contagem, vamos utilizar a coluna 'ID',
#junto com a função nunique(), que faz a contagem única de variáveis.
df.groupby('GENERO')['ID'].nunique()
```

```python
#para conseguirmos vizuaizar os valore nulos, podemos utilizar a função dropna, que é uma função binária nativa do groupby, ou seja, ela de forma
#nativa já vem aplica a função groupb como TRUE, que significa que os valores nulos evem ser retirados.
#porem quando evidenciamos ela no código com o valor FALSE, demonstra que queremos ver também a quantidade de valores nulos que temos em nossa base.
df.groupby('GENERO', dropna=False)['ID'].nunique()
```

```python
#filtrando, agrupando e contando variáveis
df[df['IDADE']>30]['NIVEL'].value_counts()
```

```python
#filtrando, agrupando e contando variáveis de forma combinada, utilizando mais de um parametro
df[(df['IDADE']>30) & (df['GENERO']=='Feminino')]['NIVEL'].value_counts()
```

```python
#criando uma tabela dinâmica com a função pivot_table do pandas
#index=coluna de parametro que será utilizaca; columns=segunda coluna de parametro que sera utilizada; values=coluna de contagem;
#aggfunc=qual função desejamos, contagem, média, mediana e etc.
pd.pivot_table(df, index='GENERO', columns='GESTOR?', values='ID', aggfunc='count')
```

```python
#outra opção para criar uma tabla dinamica com o pandas é a função crosstab
pd.crosstab(df['GENERO'], df['GESTOR?'], values=df['ID'], aggfunc='count')
```
